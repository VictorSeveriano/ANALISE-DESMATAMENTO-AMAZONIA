```python
# Lista dos estados com valores de desmatamento em cada estado de 2020 a 2023

lista_ac = [706.00, 889.00, 840.00, 601.00]
lista_ap = [24.00, 17.00, 14.00, 17.00]
lista_am = [1512.00, 2306.00, 2594.00, 1610.00]
lista_ma = [336.00, 350.00, 271.00, 306.00]
lista_ms = [1779.00, 2213.00, 1927.00, 2048.00]
lista_pa = [4899.00, 5238.00, 4162.00, 3299.00]
lista_ro = [1273.00 , 1673.00, 1480.00, 867.00]
lista_rr = [297.00, 315.00, 279.00, 284.00]
lista_to = [25.00, 37.00, 27.00, 32.00]

#Dicionário

estados = {
    "Acre": lista_ac,
    "Amapa": lista_ap,
    "Amazonas": lista_am,
    "Maranhão": lista_ma,
    "Mato Grosso": lista_ms,
    "Para": lista_pa,
    "Rondonia": lista_ro,
    "Roraima": lista_rr,
    "Tocantins": lista_to
}
```

```python
# Soma do total desmatado em cada estado de 2020 a 2023

matriz_desmatamento = [lista_ac, lista_ap, lista_am, lista_ma, lista_ms, lista_pa, lista_ro, lista_rr, lista_to]
desmatamento_total = {}

for estado, dados_estado in estados.items():
    total = sum(dados_estado)
    desmatamento_total[estado] = total

from rich import print

print("[green]Desmatamento total por estado:[/]")

for estado, total in desmatamento_total.items():
    print(f"{estado}: {total} km²")

# Calculo do total desmatado na Amazônia Legal

total_desmatamento_amazonia_legal = sum(sum(estado) for estado in matriz_desmatamento)
print(f"[yellow]Total de desmatamento na Amazônia Legal é:[/] {total_desmatamento_amazonia_legal:.2f} km²\n")
```

```python
# Calculamos a média de desmatamento de cada estado de 2020 a 2023

medias_anuais = {}

for i, j in estados.items():
    total = sum(j)
    media = total / len(j)
    medias_anuais[i] = media
for i, media in medias_anuais.items():
    print(f"[green]A média de desmatamento de [white]2020 a 2023[/] no estado de/do[/] [yellow]{i}:[/] {media:.2f} km²")
```

```python
# Porcentagem e similaridade que calculamos em proporção em relação ao total desmatado na Amazônia Legal

nomes_estados = ["Acre", "Amapá", "Amazonas", "Maranhão", "Mato Grosso", "Pará", "Rondônia", "Roraima", "Tocantins"]

for i in range(len(matriz_desmatamento)):
    estado = nomes_estados[i]
    total_estado = sum(matriz_desmatamento[i])
    porcentagem = (total_estado / total_desmatamento_amazonia_legal) * 100
    print(f"[green]A porcentagem de desmatamento de/do[/] {estado} em relação ao total desmatado na Amazônia Legal é: {porcentagem:.2f}%")
```
